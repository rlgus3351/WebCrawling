```python
# 총 리뷰 건수 수집하기
url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=74977&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=1"
res = requests.get(url, headers=h)
html = bs(res.text,'lxml')
total = html.select_one(".total > em")
# total.text  --> 24,897건이 나온다. 
# 정수형으로 바꾸려면 ,때문에 걸려서 ,를 다른 빈칸으로 변환해준다.

total_num = total.text.replace(',', "")
```

```python
num = int(total_num)
num
```

```python
# 숫자로 변환된 리뷰건수 데이터를 이용해서 페이지수 구하기
pageNum = num//10

if (num%10>=1):
    pageNum += 1
pageNum
```

```python
total_review = []
```

```python
for pageNo in range(1,pageNum+1):
    url = f"https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=74977&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={pageNo}"
    print(f"페이지 번호 : {pageNo}")
    # 리뷰 페이지를 요청한 후 리뷰데이터를 출력해보기
    res = requests.get(url, headers=h)
    html = bs(res.text,'lxml')
    
    # html을 가져와서 해당 #_filtered_ment_0~#_filtered_ment_9까지 review_data에 넣기
    try:
        for i in range(10):
            review = html.select_one(f'#_filtered_ment_{i}')
            total_review.append(review.text.strip())
    except AttributeError as err:
        print("리뷰 데이터 없음")
        
    # 페이지별로 데이터 수집하는 주기 설정
    # 2초마다 한 한번씩 실행!
    time.sleep(2)
```

```python
f = open("review_test.txt", 'w', encoding='utf-8')
for pageNo in range(101):
    url = f"https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=74977&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={pageNo+1}"
    print(f"페이지 번호 : {pageNo+1}")
    # 리뷰 페이지를 요청한 후 리뷰데이터를 출력해보기
    res = requests.get(url, headers=h)
    html = bs(res.text,'lxml')
    # html을 가져와서 해당 #_filtered_ment_0~#_filtered_ment_9까지 review_data에 넣기
    try:
        for i in range(10):
            review = html.select_one(f'#_filtered_ment_{i}')
            # total_review.append(review.text.strip())
            f.write(review.text.strip())
            f.write('\n')
    except AttributeError as err:
        print("리뷰 데이터 없음")
        
    # 페이지별로 데이터 수집하는 주기 설정
    # 2초마다 한 한번씩 실행!
    time.sleep(2)
    
f.close()
```

```python
# 문자열 데이터를 단어로 분리해주는 도구
from wordcloud import WordCloud

# 단어로 나눠어진 데이터를 시각화

import matplotlib.pyplot as plt
```

```python
# WordCloud() : 워드 클라우드 스타일 지정
# generate(문자열 데이터) : 문자열 데이터 -> 단어로 분리
wc = WordCloud(font_path="C:\Windows\Fonts\malgun.ttf", 
                background_color="white").generate(reviewData)
```

```python
plt.figure(figsize=(12,9))

# wc이미지 출력
plt.imshow(wc)
# x,y축 삭제
plt.axis('off')


# 이미지 저장
plt.savefig('wordcloud.png')
```
