#### 네이버 특정 뉴스 데이터 수집 실습1
- 뉴스 제목과 내용 수집
- HTTP 구조 -> Header(헤더)

```python
import requests
from bs4 import BeautifulSoup as bs
```

- User-Agent(브라우저 정보)


```python
# Header의 작성 이유
# Referer: 이전 페이지


h = {
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}
```

```python
url = "https://n.news.naver.com/mnews/article/003/0011704345?sid=105"
res = requests.get(url, headers=h)
```

```python
res
```

```python
page = res.text
soup = bs(page, 'lxml')
```

```python
# 뉴스 제목과 내용을 가진 요소를 선택한 후 
# 각 요소의 내용을 출력하기!
title = soup.select_one("#title_area").text
title
```

```python
content = soup.select("#dic_area")
```

```python
print(title)
print(f"{content[0].text}")
```
