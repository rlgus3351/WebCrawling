#### 영화 평점 데이터 수집
- 네이버 영화 평점 페이지에서 영화명과 평점 수집
- 날짜별로 평점 페이지를 접근해서 영화명과 평점을 수집
- 페이지를 이동하면서 수집하는 방법 학습

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

h = {
    "user-agent" :"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}
```

영화명, 평점 -> DataFrame


```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20230222"

res = requests.get(url, headers=h)
page = res.text
```

```python
test = bs(page, 'lxml')
test
```

```python
title = test.select('.tit5')
title[0].text.strip()
title_list = []
for i in title:
    title_list.append(i.text.strip())
```

```python
point = test.select('.point')
point_list = []
for i in point:
    point_list.append(i.text.strip())
```

```python
data = {"제목":title_list, "평점":point_list}
```

```python
data = {"제목":title_list, "평점":point_list}
data_df = pd.DataFrame(data)
data_df
```

#### pandas를 이용해서 날짜 데이터 생성하기

```python
# pd.date_range(시작 날짜, 마지막 날짜)

date_list = pd.date_range(start="2023-01-01", end="2023-01-31")

# pd.data_range(시작날짜, 생성할 날짜 갯수)
date_list = pd.date_range('2023-01-01',periods=31)


# Timestamp --> 문자열 반환(형식 : 년월일): %Y%m%d
date_list[0].strftime('%Y%m%d')
```

```python
dateList = []
for date in date_list:
    dateList.append(date.strftime("%Y%m%d"))
```

```python
dateList
```

#### 날짜별 영화 평점 데이터 요청하기

```python
title_list = []
point_list = []
for date in dateList:
    # url = f"https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}"
    # --> params가 필요가 없음.
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&"
    
    res = requests.get(url, headers=h, params={"date":date})
    html = bs(res.text,'lxml')
    
    point = html.select('td.point')
    title = html.select('div.tit5>a')

    # 영화 제목 수집
    for i in title:
        title_list.append(i.text.strip())
    # 영화 평점 수집
    for j in point:
        point_list.append(j.text.strip())
    
data = {"영화명" : title_list, "평점":point_list}
data_df = pd.DataFrame(data)
data_df
```
