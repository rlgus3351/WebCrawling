# G마켓 베스트 상품 데이터 수집
1. 상품명, 가격 수집
2. 수집한 정보를 엑셀파일로 저장

```python
# 모듈 가져오기
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
```

#### url 연결

```python
url = "https://corners.gmarket.co.kr/bestsellers"
driver = webdriver.Chrome()
driver.get(url)
```

#### 1. 상품명, 가격 수집

```python
item = driver.find_elements(By.CSS_SELECTOR, "a.itemname")
price = driver.find_elements(By.CSS_SELECTOR, "div.s-price > strong > span")
```

```python
item_list = []
price_list = []

for i in item:
    item_list.append(i.text)
    
for j in price:
    price_list.append(j.text.replace("원",""))
```

#### 2. 수집한 정보를 엑셀파일로 저장

```python
dict = {"제품명":item_list, "가격":price_list}
data = pd.DataFrame(dict)
data
```

```python
data.to_excel('./G마켓 베스트 상품 데이터.xlsx', index=False)
```

```python
url = "https://corners.gmarket.co.kr/bestsellers"
driver = webdriver.Chrome()
driver.get(url)
for i in range(12):
    driver.find_element(By.CSS_SELECTOR,f"li.group{i+1}").click()
    time.sleep(1)
    cate_name_list = driver.find_elements(By.CSS_SELECTOR, "a.itemname")
    cate_price_list = driver.find_elements(By.CSS_SELECTOR, "div.s-price > strong > span")
    for x in range(len(cate_name_list)):
        cate_name.append(cate_name_list[x].text)
        cate_price.append(cate_price_list[x].text)
        categori = driver.find_element(By.CSS_SELECTOR,f"li.group{i+1} > a").text
        cate_list.append(categori)
    dict = {"상품명":cate_name, "가격":cate_price, "카테고리": cate_list}
        
```

- (실습) 카테고리를 1초마다 클릭해서 넘어가는 로직 구현하기!

```python
# 모듈 가져오기
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
```

```python
url = "https://corners.gmarket.co.kr/bestsellers"
driver = webdriver.Chrome()
driver.get(url)
```

```python
# ul>li>a 형태
groupList = driver.find_elements(By.CSS_SELECTOR, '#categoryTabG > li > a')
groupList
len(groupList)
```

```python
# 첫번째 요소 접근 글자 추출
groupList[0].text
```

```python
pdNameList2 = []
pdPriceList2 = []
groupList2 = []

# ALL이 아니라, 패션의류부터 시작 
# 카테고리가 12개 고정이라 범위를 1~12까지 설정해준다.
for i in range(1, 13):
    
    # 위에 groupList에 담았던 변수들이 페이지를 돌아다니면서 요소상태가 변해서 오류가 나서 for문 안에 넣어준다.
    group = driver.find_elements(By.CSS_SELECTOR, '#categoryTabG > li > a')
    
    # click()으로 페이지가 넘어가기 전에 그룹데이터를 변수에 저장하고
    # 각 상품에 그룹 데이터를 매핑하기 위한 데이터
    groupNm = group[i].text
    
    group[i].click()
    time.sleep(1)
    # 그룹리스트별 a를 클릭
    
    # 그룹별 상품 데이터 수집
    pdNameList = driver.find_elements(By.CSS_SELECTOR, "a.itemname")
    pdPriceList = driver.find_elements(By.CSS_SELECTOR, 'span.for-a11y + strong > span')
    print("-"*50)
    
    for j in range(len(pdNameList)):
        pdNameList2.append(pdNameList[j].text)
        pdPriceList2.append(pdPriceList[j].text)
        
        # 그룹에 대한 데이터를 넣어주기 위해
        groupList2.append(groupNm)
    
```

```python
# 리스트 개수 파악하기!
len(pdNameList2), len(pdPriceList2), len(groupList2)
```

```python
gmarket_dic = {'상품명': pdNameList2 , "가격": pdPriceList2, '그룹': groupList2}
gmarket_df = pd.DataFrame(gmarket_dic)
gmarket_df
```
